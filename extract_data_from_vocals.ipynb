```python
import numpy as np
import matplotlib.pyplot as plt
import os
import librosa
from tqdm import tqdm
```

```python
main_path='/content/drive/MyDrive/output_data'
```

```python
all_data=[]
for i in tqdm(os.listdir(main_path)):
  file_name=main_path+'/'+i+'/vocals.wav'
  audio_data, sample_rate = librosa.load(file_name, sr=None)
  all_data.append(audio_data[np.abs(audio_data)>0.0004])
```

```python
min_l=np.inf
for i in all_data:
  if len(i)<min_l:
    min_l=len(i)
```
